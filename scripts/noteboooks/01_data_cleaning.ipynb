```python
# Mapeamento completo dos dados categóricos - ENEM 2023 Baseado no dicionário de microdados

mapeamentos = {
    # === DADOS DEMOGRÁFICOS (APENAS NÃO-ORDINAIS) ===
    'tipo_sexo': {
        'M': 'Masculino',
        'F': 'Feminino'
    },
    
    'tipo_cor_raca': {
        0: 'Não declarado',
        1: 'Branca',
        2: 'Preta',
        3: 'Parda',
        4: 'Amarela',
        5: 'Indígena',
        6: 'Não dispõe da informação'
    },
    
    'tipo_nacionalidade': {
        0: 'Não informado',
        1: 'Brasileiro(a)',
        2: 'Brasileiro(a) Naturalizado(a)',
        3: 'Estrangeiro(a)',
        4: 'Brasileiro(a) Nato(a), nascido(a) no exterior'
    },
    
    # === DADOS EDUCACIONAIS === # Aqui vai uma reodenação ordinal dessa variavel que utilizarei para correlações futuras
    'tipo_conclusao': {
        1: 4,
        2: 2,
        3: 3,
        4: 1
    },
    
    'tipo_escola': {
        1: 'Não Respondeu',
        2: 'Pública',
        3: 'Privada'
    },
    
    'tipo_ensino': {
        1: 'Ensino Regular',
        2: 'Educação Especial - Modalidade Substitutiva'
    },
    
    'tipo_treineiro': {
        0: 'Não é treineiro',
        1: 'É treineiro'
    },
    
    # === LOCALIZAÇÃO ===
    'zona_escola': {
        1: 'Urbana',
        2: 'Rural'
    },
    
    'status_escola': {
        1: 'Em atividade',
        2: 'Paralisada',
        3: 'Extinta',
        4: 'Escola extinta em anos anteriores'
    },
    
    # === DADOS DA PROVA ===
    'tipo_presenca_cn': {
        0: 'Faltou à prova',
        1: 'Presente na prova',
        2: 'Eliminado na prova'
    },
    
    'tipo_presenca_ch': {
        0: 'Faltou à prova',
        1: 'Presente na prova',
        2: 'Eliminado na prova'
    },
    
    'tipo_presenca_lc': {
        0: 'Faltou à prova',
        1: 'Presente na prova',
        2: 'Eliminado na prova'
    },
    
    'tipo_presenca_mt': {
        0: 'Faltou à prova',
        1: 'Presente na prova',
        2: 'Eliminado na prova'
    },
    
    'lingua_estrangeira': {
        0: 'Inglês',
        1: 'Espanhol'
    },
    
    'status_redacao': {
        1: 'Sem problemas',
        2: 'Anulada',
        3: 'Cópia Texto Motivador',
        4: 'Em Branco',
        6: 'Fuga ao tema',
        7: 'Não atendimento ao tipo textual',
        8: 'Texto insuficiente',
        9: 'Parte desconectada'
    }
}

# Função para aplicar os mapeamentos
def aplicar_mapeamentos(df, mapeamentos):
    """
    Aplica os mapeamentos categóricos ao DataFrame com uma validação se a coluna existe no dataframe
    """
    df_mapped = df.copy()
    
    for coluna, mapeamento in mapeamentos.items():
        if coluna in df_mapped.columns:
            df_mapped[coluna] = df_mapped[coluna].map(mapeamento)
            print(f"✅ Mapeamento aplicado para: {coluna}")
        else:
            print(f"⚠️ Coluna não encontrada: {coluna}")
    
    return df_mapped
```

```python
# Aplicando função para mapear os dados
mapped_data = aplicar_mapeamentos(raw_data, mapeamentos)
```

```python
# Dando uma olhada nos dados
mapped_data.head(7)
```

```python
# Criando coluna de situacao presenca para verificar se o candidato teve nas provas objetivas a (presença em todas as provas, ausencia total, eliminação total ou outro)

mapped_data['situacao_presenca'] = np.where(
    (mapped_data['tipo_presenca_cn'] == 'Presente na prova') &
    (mapped_data['tipo_presenca_ch'] == 'Presente na prova') &
    (mapped_data['tipo_presenca_lc'] == 'Presente na prova') &
    (mapped_data['tipo_presenca_mt'] == 'Presente na prova'),
    'presenca_total',
    np.where(
        (mapped_data['tipo_presenca_cn'] == 'Faltou à prova') &
        (mapped_data['tipo_presenca_ch'] == 'Faltou à prova') &
        (mapped_data['tipo_presenca_lc'] == 'Faltou à prova') &
        (mapped_data['tipo_presenca_mt'] == 'Faltou à prova'),
        'ausencia_total',
        np.where(
            (mapped_data['tipo_presenca_cn'] == 'Eliminado na prova') &
            (mapped_data['tipo_presenca_ch'] == 'Eliminado na prova') &
            (mapped_data['tipo_presenca_lc'] == 'Eliminado na prova') &
            (mapped_data['tipo_presenca_mt'] == 'Eliminado na prova'),
            'eliminacao_total',
            np.where(
                (mapped_data['tipo_presenca_cn'] == 'Presente na prova') |
                (mapped_data['tipo_presenca_ch'] == 'Presente na prova') |
                (mapped_data['tipo_presenca_lc'] == 'Presente na prova') |
                (mapped_data['tipo_presenca_mt'] == 'Presente na prova'),
                'presenca_parcial',
                'outro'  # Caso padrão para situações mistas
        )  # Caso padrão para situações mistas
        )
    )
)
```

```python
# Verificando dist relativa da situacao_presenca e fitlrando somente alunos com participacao total

print((mapped_data['situacao_presenca'].value_counts() / mapped_data['situacao_presenca'].value_counts().sum()) * 100, '\n')

mapped_data = mapped_data.query(
    "situacao_presenca == 'presenca_total'"
)

print((mapped_data['situacao_presenca'].value_counts() / mapped_data['situacao_presenca'].value_counts().sum()) * 100)
```

```python
# dropando colunas de presenca que já são obsoletas
mapped_data.drop(['tipo_presenca_cn', 'tipo_presenca_ch', 'tipo_presenca_lc', 'tipo_presenca_mt', 'situacao_presenca'], axis=1,inplace=True)

mapped_data.head(7)
```

*TRATANDO NULOS*

```python
# verificando percentual de nulos por coluna
print((mapped_data.isna().mean() * 100).sort_values(ascending=False))
```

```python
# Dropando colunas com alto indece de nulos a partir de um for para identificar as colunas com indece maior que 50% de nulos

# Dicionario com as colunas e os respectivos indices de nulos
nulos_dict  = (mapped_data.isna().mean() * 100).sort_values(ascending=False).to_dict()

# Verificando o dicionario
print(nulos_dict, '\n')

# Criando lista vazia a ser preenchida com as colunas com alto indice de nulos para o drop
colunas_defasadas = []

# For para preencher lista defasadas de acordo com o indice de nulos
for coluna, indice in nulos_dict.items():
    if indice > 50:
        colunas_defasadas.append(coluna)

#Verificando resultado do preenchimento
print(colunas_defasadas, '\n')
```

```python
# Dropando colunas com alto indece de nulos 
mapped_data = mapped_data.drop(colunas_defasadas, axis=1)
```
