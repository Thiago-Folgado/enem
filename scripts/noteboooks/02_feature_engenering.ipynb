```python
# Criando função para mapear as colunas de variaveis ordinais para ordinais numericas

def mapear_var_ordinais(df, mapeamento):

    df_mapped = df.copy()

    for coluna, mapeamento in mapeamento.items():
        if coluna in df.columns:
           df_mapped[coluna] = df_mapped[coluna].map(mapeamento)
           print(f"✅ Mapeamento aplicado para: {coluna}")
        else:
            print(f"⚠️ Coluna não encontrada: {coluna}")
    return df_mapped
```

```python
# Criando função para criar uma variavel de "Faixa renda per capita" de acordo com a variavel criada de "Renda per capita", as faixas auxiliaram nas analises posteriores

def criar_faixas_renda_per_capita(df):
    """
    Cria diferentes tipos de faixas para renda per capita
    
    Parameters:
    df: DataFrame com coluna 'renda_per_capita'
    metodo:'salario_minimo': Baseado no salário mínimo 2023 (R$ 1.320)
    Returns:
    DataFrame com novas colunas de faixas
    """
        
    # Salário mínimo 2023 = R$ 1.320
    sm_2023 = 1320
    
    bins_sm = [0, sm_2023*0.25, sm_2023*0.5, sm_2023*0.75, 
                sm_2023*1, sm_2023*1.5, sm_2023*2, sm_2023*3, 
                sm_2023*5, float('inf')]
    
    labels_sm = [
        1,            #'A - Extrema pobreza (até 1/4 SM)',     até R$ 330
        2,            #'B - Pobreza (1/4 a 1/2 SM)',           R$ 330 - 660
        3,            #'C - Baixa renda (1/2 a 3/4 SM)',       R$ 660 - 990
        4,            #'D - Renda baixa (3/4 a 1 SM)',         R$ 990 - 1.320
        5,            #'E - Classe baixa (1 a 1,5 SM)',        R$ 1.320 - 1.980
        6,            #'F - Classe média baixa (1,5 a 2 SM)',  R$ 1.980 - 2.640
        7,            #'G - Classe média (2 a 3 SM)',          R$ 2.640 - 3.960
        8,            #'H - Classe média alta (3 a 5 SM)',     R$ 3.960 - 6.600
        9,            #'I - Classe alta (acima 5 SM)'          acima R$ 6.600
    ]
    
    df['faixa_renda_per_capita'] = pd.cut(
        df['renda_per_capita'], 
        bins=bins_sm, 
        labels=labels_sm, 
        include_lowest=True
    )
```

```python
# Aplicando função para criar coluna de "Faixa renda per capita"

criar_faixas_renda_per_capita(mapped_data)
```
